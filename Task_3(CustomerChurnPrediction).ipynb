```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/content/drive/MyDrive/CodSoft/Churn_Modelling.csv")
df.head()
```

```python
df.info()
```

```python
df["Geography"].unique()
```

```python
df.describe()
```

```python
plt.bar(df['Gender'].value_counts().index, df['Gender'].value_counts().values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
region = df['Geography'].value_counts().reset_index()
region.columns = ['Country', 'Count']
plt.figure(figsize = (5,5))
ax = sns.barplot(x=region['Country'], y = region['Count'])
for i in ax.containers:
  ax.bar_label(i,)
```

```python
df.drop(columns= ['RowNumber', 'CustomerId', 'Surname'], inplace = True)
df.head()
```

```python
labelencoder = LabelEncoder()
df['Geography'] = labelencoder.fit_transform(df['Geography'])
df['Gender'] = labelencoder.fit_transform(df['Gender'])
df.head()
```

```python
df.dtypes
```

```python
x = df.drop(columns = 'Exited')
y = df['Exited']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
y_preds = model.predict(x_test)
y_preds
```

```python
len(y_preds)
```

```python
accuracy = accuracy_score(y_test, y_preds)
accuracy
```

```python
y_prob = model.predict_proba(x_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr,color="red", label=f'AUC = {auc(fpr, tpr):.2f}')
plt.plot([0,1],[0,1], color="blue", lw=2);
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
y_preds = model.predict(x_test)
y_preds
```

```python
accuracy = accuracy_score(y_test, y_preds)
accuracy
```

```python
y_prob = model.predict_proba(x_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr,color="red", label=f'AUC = {auc(fpr, tpr):.2f}')
plt.plot([0,1],[0,1], color="blue", lw=2);
```

```python
model = GradientBoostingClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
y_preds = model.predict(x_test)
y_preds
```

```python
accuracy = accuracy_score(y_test, y_preds)
accuracy
```

```python
y_prob = model.predict_proba(x_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr,color="red", label=f'AUC = {auc(fpr, tpr):.2f}')
plt.plot([0,1],[0,1], color="blue", lw=2);
```

```python
model_names = ["Random Forest", "Logistic Regression", "GradientBoosting"]
```

```python
acc_list = []
```

```python
for model in [RandomForestClassifier(), LogisticRegression(), GradientBoostingClassifier()]:
    model.fit(x_train, y_train)
    y_preds = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_preds)
    acc_list.append(accuracy)
```

```python
model_result = pd.DataFrame({'Model': model_names, 'Accuracy': acc_list})
model_result
```

```python
sns.barplot(x= "Model", y= "Accuracy", data = model_result);
```
